### Reusing provided python code

```python
# Importing more modules
from __future__ import print_function
import argparse
import json
import pprint
import requests
import sys
import urllib

# This client code can run on Python 2.x or 3.x.  Your imports can be
# simpler if you only need one of those.
try:
    # For Python 3.0 and later
    from urllib.error import HTTPError
    from urllib.parse import quote
    from urllib.parse import urlencode
except ImportError:
    # Fall back to Python 2's urllib2 and urllib
    from urllib2 import HTTPError
    from urllib import quote
    from urllib import urlencode

# Yelp Fusion no longer uses OAuth as of December 7, 2017.
# You no longer need to provide Client ID to fetch Data
# It now uses private keys to authenticate requests (API Key)
# You can find it on
# https://www.yelp.com/developers/v3/manage_app
API_KEY= yelpkeys.api_key


# API constants, you shouldn't have to change these.
API_HOST = 'https://api.yelp.com'
SEARCH_PATH = '/v3/businesses/search'
BUSINESS_PATH = '/v3/businesses/'  # Business ID will come after slash.

# Assigning a search limit
SEARCH_LIMIT = 20


# Function needed for search function
def request(host, path, api_key, url_params=None):
    """Given your API_KEY, send a GET request to the API.
    Args:
        host (str): The domain host of the API.
        path (str): The path of the API after the domain.
        API_KEY (str): Your API Key.
        url_params (dict): An optional set of query parameters in the request.
    Returns:
        dict: The JSON response from the request.
    Raises:
        HTTPError: An error occurs from the HTTP request.
    """
    url_params = url_params or {}
    url = '{0}{1}'.format(host, quote(path.encode('utf8')))
    headers = {
        'Authorization': 'Bearer %s' % api_key,
    }

    print(u'Querying {0} ...'.format(url))

    response = requests.request('GET', url, headers=headers, params=url_params)

    return response.json()


# Function for searching businesses
def search(api_key, term, location):
    """Query the Search API by a search term and location.
    Args:
        term (str): The search term passed to the API.
        location (str): The search location passed to the API.
    Returns:
        dict: The JSON response from the request.
    """

    url_params = {
        'term': term.replace(' ', '+'),
        'location': location.replace(' ', '+'),
        'limit': SEARCH_LIMIT
    }
    return request(API_HOST, SEARCH_PATH, api_key, url_params=url_params)
```

```python
# Calling the search function and naming the json file as 'response'
response = search(API_KEY,'Barbershop', 'Montclair, NJ')
```

```python
# Printing response to make sure everything worked correctly
print(response)
```

### Exploring the JSON file

```python
# Creating a tally in order to keep track of reviews
positive = 0
negative = 0
neutral = 0

for business in response['businesses']:
    print(response)
    print('Business Name: ' + business['name'])
    print('Rating: ' + str(business['rating']))
    print('Review Count: ' + str(business['review_count']))
    print('Reviews: ')
    BUS_REVIEW='/v3/businesses/{}/reviews'.format(business['id'])
    #print('https://api.yelp.com/v3/businesses/{}/reviews?limit=20&sort_by=yelp_sort'.format(business['id']))
    reviews = request(API_HOST,BUS_REVIEW, API_KEY)
    for review in reviews['reviews']:
        print(' - ' + review['text'])
        # Perform sentiment analysis on the review using TextBlob
        blob = TextBlob(review['text'])
        sentiment = blob.sentiment.polarity
        if sentiment > 0:
            positive += 1
            print('    Sentiment: Positive')
        elif sentiment < 0:
            negative += 1
            print('    Sentiment: Negative')
        else:
            print('    Sentiment: Neutral')
            neutral += 1
        print('\n================\n\n\n')
```

```python
# Printing tally of reviews
print(f"Positive Reviews: {positive}\nNegative Reviews: {negative}\nNeutral Reviews: {neutral}")
```

### Using provided code to create a donut chart

## Exploring the data using NaiveBayesAnalyzer

```python
# Importing module
from textblob.sentiments import NaiveBayesAnalyzer

pos = 0
neg = 0
neut = 0

# Creating a loop to go through each business
for business in response['businesses']:
    print(response)
    print('Business Name: ' + business['name'])
    print('Rating: ' + str(business['rating']))
    print('Review Count: ' + str(business['review_count']))
    print('Reviews: ')
    BUS_REVIEW='/v3/businesses/{}/reviews'.format(business['id'])
    #print('https://api.yelp.com/v3/businesses/{}/reviews?limit=20&sort_by=yelp_sort'.format(business['id']))
    reviews = request(API_HOST,BUS_REVIEW, API_KEY)
    for review in reviews['reviews']:
        print(' - ' + review['text'])
        # Perform sentiment analysis on the review using NaiveBayesAnalyzer
        blob = TextBlob(review['text'], analyzer = NaiveBayesAnalyzer())
        sentiment = blob.sentiment.classification
        if sentiment == 'pos':
            pos += 1
            print('    Sentiment: Positive')
        elif sentiment == 'neg':
            neg += 1
            print('    Sentiment: Negative')
        else:
            print('    Sentiment: Neutral')
            neut += 1
        print('\n================\n\n\n')
```

```python
# Printing tally of reviews
print(f"Positive Reviews: {pos}\nNegative Reviews: {neg}\nNeutral Reviews: {neut}")
```

## Exploring most common words in reviews

```python
# Importing stopwords from nltk
from nltk.corpus import stopwords

# assigning stopwords to stops list
stops = stopwords.words('english')

# Printing list
stops
```

### Tallying up words from reviews

```python
# Creating a dictionary that will house the words and their tally
word_dict = {}

# Creating a function 'word_tally' that creates a dictionary of words used as its keys and the tally as its value
def word_tally(x):
    for word in x:
        if word in word_dict.keys():
            word_dict[word] += 1
        if word not in word_dict.keys():
            word_dict[word.lower()] = 1

# Creating a loop to go through each review
for business in response['businesses']:
    BUS_REVIEW='/v3/businesses/{}/reviews'.format(business['id'])
    reviews = request(API_HOST,BUS_REVIEW, API_KEY)
    for review in reviews['reviews']:
        # splititing the reviews into lists and running the list through the word_tally function
        word_tally(review['text'].split())

# Printing word_dict to make sure it worked
print(word_dict)
```

### Sorting the words via their tally

```python
# Sorting the words_dict by its values in descending order. Used old code from project 2
sorted_words_dict = dict(sorted(word_dict.items(), key=lambda x: x[1], reverse=True))

# Printing updated 'sorted_runtimes_dict' to check for errors
print(sorted_words_dict)
print("\n")
```

### Removing stop words

```python
# Iterating over a copy of the dictionary keys as this gets rid of the error as solved using ChatGPT
for word in list(sorted_words_dict.keys()):
    if word in stops:
        del sorted_words_dict[word]

# Printing to check for errors
print(sorted_words_dict)
```
